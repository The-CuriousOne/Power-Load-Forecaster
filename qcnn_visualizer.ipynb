```python
import torch
import matplotlib.pyplot as plt

# Load the .pth file
checkpoint = torch.load("qcnn_model.pth")
```

```python
# Extract training and validation history from logs
train_loss = checkpoint['train_losses']
train_acc = checkpoint['train_accuracies']
val_loss = checkpoint['val_losses']
val_acc = checkpoint['val_accuracies']
```

```python
# Verify the contents
print("Training Loss:", train_loss)
print("Training Accuracy:", train_acc)
print("Validation Loss:", val_loss)
print("Validation Accuracy:", val_acc)
print("Number of epochs:", len(train_loss))
```

```python
# Extract the number of epochs
epochs = range(1, len(train_loss) + 1)
```

```python
# Plot Training Loss over Epochs
plt.figure(figsize=(10, 5))
plt.plot(epochs, train_loss, label="Training Loss")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Training Accuracy over Epochs
plt.figure(figsize=(10, 5))
plt.plot(epochs, train_acc, label="Training Accuracy", color='orange')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.title('Training Accuracy over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Validation Loss over Epochs
plt.figure(figsize=(10, 5))
plt.plot(epochs, val_loss, label="Validation Loss", color='red')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Validation Loss over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Validation Accuracy over Epochs
plt.figure(figsize=(10, 5))
plt.plot(epochs, val_acc, label="Validation Accuracy", color='green')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.title('Validation Accuracy over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Accuracy vs Epochs (combined plot)
plt.figure(figsize=(10, 5))
plt.plot(epochs, train_acc, label="Training Accuracy", color='orange')
plt.plot(epochs, val_acc, label="Validation Accuracy", color='green')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.title('Model Accuracy over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```
